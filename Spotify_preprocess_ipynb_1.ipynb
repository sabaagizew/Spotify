```python
categorical_columns = df.select_dtypes(include=['object']).columns
print("\nUnique Values in Categorical Columns:")
for col in categorical_columns:
    print(f"{col}: {df[col].unique()[:10]}")
```

```python
# Check for outliers using IQR for numerical columns
numerical_columns = df.select_dtypes(include=['number']).columns
print("\nOutlier Counts per Numerical Column:")
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outlier_count = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))].shape[0]
    print(f"{col}: {outlier_count} outliers")
```

```python

# Remove duplicate rows
df_cleaned = df.drop_duplicates()
```

```python
# Check for invalid 'year' values (assuming valid years are between 1900 and 2025)
valid_years = (df_cleaned['year'] >= 1900) & (df_cleaned['year'] <= 2025)
df_cleaned = df_cleaned[valid_years]
```

```python
# Handling outliers: Removing extreme values using IQR (only for relevant numerical columns)
numerical_columns = df_cleaned.select_dtypes(include=['number']).columns
for col in numerical_columns:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
```

```python
# Save cleaned dataset
cleaned_file_path = "spotify_tracks_cleaned.csv"
df_cleaned.to_csv(cleaned_file_path, index=False)

print(f"Cleaned dataset saved as {cleaned_file_path}")
```
